```python
import pandas as pd
data = [1,2,3,4,5]

# convert list into panda series
#to assign index to the values in data

series = pd.Series(data ,index=[101,102,103,104,105])
print(series)
print(series[series > 3]) # this only works when the values in data is integer
```

```python
weeks = {
    "day 1" : "monday" ,
    "day 2" : "tuesday" ,
    "day 3" : "wednesday" ,
    "day 4" : "thursday" ,
    "day 5" : "friday" ,
    "day 6" : "saturday" ,
    "day 7" : "sunday"
}


series1 = pd.Series(weeks)  # now the keys in the dictonary will become the indexes of the value
print(series1)
print("first elemented in dictoanry")
print(series1.iloc[0]) # acces using the orginal index values 
print(series1.loc["day 2"]) # acces using the key index values
```

```python
# -----------------------------------------------------
# ⿡ Series from a list with filtering
# -----------------------------------------------------
# Task:
# - Create a Series from the list [10, 20, 30, 40, 50]
# - Use loc[] to access the element with index 3
# - Use iloc[] to access the second element
# - Filter and print elements greater than 25


s = [10,20,30,40,50]
se = pd.Series(s,index=[101,102,103,104,105])
print(se.iloc[1])
print(se.loc[103])
print(se[se>25])
```

```python
# -----------------------------------------------------
# ⿢ Series from a dictionary
# -----------------------------------------------------
# Task:
# - Create a Series from the dictionary: {"A": 5, "B": 15, "C": 25, "D": 35}
# - Access the value of key "B" using loc[]
# - Access the first element using iloc[]
# - Filter and print values greater than 20


s1 = {
    "A" : 5 ,
    "B" : 15,
    "C" : 25,
    "D" : 35
 }

s2=pd.Series(s1)
print(s2.loc["B"])
print(s2.iloc[0])
print(s2[s2>20])
```

```python
# -----------------------------------------------------
# ⿣ DataFrame from a dictionary
# -----------------------------------------------------
# Task:
# - Create a DataFrame with columns: "Name", "Age", "City"
#   Data: ["Alice", "Bob", "Charlie"], [25, 30, 35], ["Delhi", "Mumbai", "Chennai"]
# - Print the entire DataFrame
# - Print the row at index 1 using loc[]
# - Print the first row using iloc[]

t = {
    "name" :["alice","bob","charlie"],
    "age" : [25,30,35],
    "city" : ["delhi","mumbai","chennai"]

}
t1 = pd.DataFrame(t)
print(t1)
print(t1.loc[1])
print(t1.iloc[0])
```

```python
# -----------------------------------------------------
# ⿤ Conditional filtering in DataFrame
# -----------------------------------------------------
# Task:
# - Using the above DataFrame, filter rows where Age > 28
# - Filter rows where City is "Delhi"
# - Print the filtered DataFrames

print(t1[t1["age"]>28])
print(t1[t1["city"] == "delhi"])
print()
```
